```python
import pandas as pd
import matplotlib.pyplot as plt
# 使DataFrame在jupyter notebook中输出不换行
pd.set_option('display.expand_frame_repr', False)
```

```python
# 数据集导入
df = pd.read_csv('../user_action_after_preprocess.csv')
print('数据集数据框：\n', df)
```

```python
# 计算不同类型行为的个数
total_actions_num = df.shape[0]
action_1_num = (df['behavior_type'] == 1).sum()
action_2_num = (df['behavior_type'] == 2).sum()
action_3_num = (df['behavior_type'] == 3).sum()
action_4_num = (df['behavior_type'] == 4).sum()
print('用户行为总数为：', total_actions_num)
print('行为"1"个数为：', action_1_num)
print('行为"2"个数为：', action_2_num)
print('行为"3"个数为：', action_3_num)
print('行为"4"个数为：', action_4_num)

# 绘制饼图
radios = [action_1_num/total_actions_num,
          action_2_num/total_actions_num,
          action_3_num/total_actions_num,
          action_4_num/total_actions_num]
labels = ['browse', 'collect', 'add_shopping_cart', 'buy']
plt.pie(radios, labels=labels, autopct='%1.2f%%', startangle=45, pctdistance=0.98) # 绘制饼图
plt.title('User action pie chart') # 设置饼图标题
plt.legend(bbox_to_anchor=(1, 0.45)) # 设置图例位置
plt.show() # 展示饼图
```
