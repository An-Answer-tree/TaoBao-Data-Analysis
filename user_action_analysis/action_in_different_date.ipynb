```python
import pandas as pd
import matplotlib.pyplot as plt
# 使DataFrame在jupyter notebook中输出不换行
pd.set_option('display.expand_frame_repr', False)
```

```python
# 数据集导入
df = pd.read_csv('../user_action_after_preprocess.csv')
print('数据集数据框：\n', df)
```

```python
# 绘制30天，action_1的变化情况
df_action_1 = df[df['behavior_type'] == 1] 
print('只有action_1的数据框：\n', df_action_1, '\n\n')

# 绘制柱状图
value_counts_1 = df_action_1['date'].value_counts()
value_counts_1.sort_index(inplace=True) # 将series按index排序
value_counts_1.plot(kind='bar', rot=0, color='skyblue') # 绘bar图
value_counts_1.plot(kind='line', marker='o', color='red', linestyle='-', linewidth=2) # 绘line图

# 设置横坐标标签斜着显示
plt.xticks(rotation=60, ha='right')

# 添加图表标题和坐标轴标签
plt.title('Count of action_browse Each date') # 设置标题
plt.xlabel('Date') # 设置横坐标名称
plt.ylabel('Count') #  设置纵坐标名称

# 显示图表
plt.show()
```
